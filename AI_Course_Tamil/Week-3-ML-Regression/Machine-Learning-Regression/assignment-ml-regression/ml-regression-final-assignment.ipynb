```python
# Problem Statement
# Identify charges for insurance based on the metrics provided.
# After looking into the dataset its numerical data. 
# Both input and outputs are available.
# Output label is numerical.
###################################################
# Domain - MACHINE LEARNING
# Supervised Learning
# Regression
###################################################
```

```python
import pandas as pd
```

```python
dataset = pd.read_csv("insurance_pre.csv")
dataset
```

```python
# Dataset has 1338 Rows X 6 Columns
# We have to convert the following nominal fields to numerical using one-hot encoding.
# The nominal fields are smoker, sex.
```

```python
# Preprocessing
dataset = pd.get_dummies(dataset, dtype="int64", drop_first=True)
dataset
```

```python
#  Display Columns
dataset.columns
```

```python
# Identify independent variables
independent = dataset[["age", "bmi", "children", "sex_male", "smoker_yes"]]
independent
```

```python
# Identify dependent variable
dependent = dataset[["charges"]]
dependent
```

```python
# Split training and test set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(independent, dependent, test_size = 0.30, random_state = 0)
```

```python
## Multiple Linear Regression
# Model Creation and Evaluation
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import ParameterGrid
from sklearn.metrics import r2_score
import pandas as pd

# Define the hyperparameter grid
param_grid = {
    "fit_intercept": [True, False],
    "copy_X": [True, False],
    "positive": [True, False],
    "tol": [1e-3, 1e-6],
}

results = []

# Iterate through all hyperparameter combinations
for params in ParameterGrid(param_grid):

    # Create the model
    model = LinearRegression(**params)

    # Train the model
    model.fit(x_train, y_train)

    # Predict
    y_pred = model.predict(x_test)

    # Calculate R2 Score
    score = r2_score(y_test, y_pred)

    # Store the results
    results.append({
        "Algorithm": "Multiple Linear Regression",
        "fit_intercept": params["fit_intercept"],
        "copy_X": params["copy_X"],
        "positive": params["positive"],
        "tol": params["tol"],
        "R2 Score": score
    })

# Convert to DataFrame
results_df = pd.DataFrame(results)

# Sort by R2 Score
results_df = results_df.sort_values(
    by="R2 Score",
    ascending=False
).reset_index(drop=True)

# Display the results
print(results_df)

# Optional: Save to CSV
results_df.to_csv("LinearRegression_Hyperparameter_Results.csv", index=False)
```

```python
# Import the Decision Tree Regression algorithm
from sklearn.tree import DecisionTreeRegressor

# Import ParameterGrid to generate all possible combinations of hyperparameters
from sklearn.model_selection import ParameterGrid

# Import R2 Score to evaluate the model performance
from sklearn.metrics import r2_score

# Import Pandas to store and display the results in a DataFrame
import pandas as pd


# Define the hyperparameter combinations to be tested
param_grid = {
    "criterion": ["squared_error", "friedman_mse"],
    "splitter": ["best", "random"],
    "max_depth": [5, 10, 15],
    "min_samples_split": [2, 5],
    "min_samples_leaf": [1, 2],
    "max_features": [None, "sqrt"]
}

# Create an empty list to store the results of each iteration
results = []

# Loop through every hyperparameter combination
for params in ParameterGrid(param_grid):

    # Create the Decision Tree model using the current hyperparameter combination
    model = DecisionTreeRegressor(
        random_state=42,
        **params
    )

    # Train the model using the training dataset
    model.fit(x_train, y_train)

    # Predict the target values using the test dataset
    y_pred = model.predict(x_test)

    # Calculate the R2 Score for the current model
    score = r2_score(y_test, y_pred)

    # Store the algorithm name, hyperparameters and R2 Score
    results.append({
        "Algorithm": "Decision Tree Regression",
        **params,
        "R2 Score": score
    })

# Convert the results list into a Pandas DataFrame
results_df = pd.DataFrame(results)

# Sort the DataFrame by R2 Score in descending order
results_df = results_df.sort_values(
    by="R2 Score",
    ascending=False
).reset_index(drop=True)

# Display all the hyperparameter combinations and their R2 Scores
print(results_df)

# Save the DataFrame to a CSV file
results_df.to_csv("DecisionTree_Hyperparameter_Results.csv", index=False)

# Display a success message
print("Results saved successfully!")
```

```python
## Support Vector Machine - Regression

# Import the Support Vector Regression algorithm
from sklearn.svm import SVR

# Import StandardScaler to scale the features
from sklearn.preprocessing import StandardScaler

# Import ParameterGrid to generate all possible hyperparameter combinations
from sklearn.model_selection import ParameterGrid

# Import R2 Score to evaluate the model
from sklearn.metrics import r2_score

# Import Pandas to store the results
import pandas as pd

# --------------------------------------------------
# Feature Scaling (Required for SVR)
# --------------------------------------------------

# Create the scaler
scaler = StandardScaler()

# Fit on training data and transform
x_train_scaled = scaler.fit_transform(x_train)

# Transform the test data using the same scaler
x_test_scaled = scaler.transform(x_test)

# --------------------------------------------------
# Hyperparameter Grids
# --------------------------------------------------

param_grids = [

    # Linear Kernel (6 combinations)
    {
        "kernel": ["linear"],
        "C": [0.1, 1, 10],
        "epsilon": [0.1, 0.5]
    },

    # RBF Kernel (24 combinations)
    {
        "kernel": ["rbf"],
        "C": [0.1, 1, 10],
        "gamma": ["scale", "auto", 0.01, 0.1],
        "epsilon": [0.1, 0.5]
    },

    # Polynomial Kernel (24 combinations)
    {
        "kernel": ["poly"],
        "C": [1, 10],
        "gamma": ["scale", "auto"],
        "degree": [2, 3, 4],
        "epsilon": [0.1, 0.5]
    },

    # Sigmoid Kernel (16 combinations)
    {
        "kernel": ["sigmoid"],
        "C": [1, 10],
        "gamma": ["scale", "auto"],
        "epsilon": [0.1, 0.5],
        "coef0": [0, 1]
    }

]

# Empty list to store results
results = []

# --------------------------------------------------
# Hyperparameter Tuning
# --------------------------------------------------

for param_grid in param_grids:

    # Generate all combinations for the current kernel
    for params in ParameterGrid(param_grid):

        # Create the model
        model = SVR(**params)

        # Train the model
        model.fit(x_train_scaled, y_train)

        # Predict on the test dataset
        y_pred = model.predict(x_test_scaled)

        # Calculate the R2 Score
        score = r2_score(y_test, y_pred)

        # Store the results
        results.append({
            "Algorithm": "Support Vector Regression",
            **params,
            "R2 Score": score
        })

# --------------------------------------------------
# Convert results into DataFrame
# --------------------------------------------------

results_df = pd.DataFrame(results)

# Arrange columns in a consistent order
columns = [
    "Algorithm",
    "kernel",
    "C",
    "gamma",
    "degree",
    "coef0",
    "epsilon",
    "R2 Score"
]

# Add missing columns (for kernels where they are not applicable)
for col in columns:
    if col not in results_df.columns:
        results_df[col] = None

results_df = results_df[columns]

# Sort by R2 Score
results_df = results_df.sort_values(
    by="R2 Score",
    ascending=False
).reset_index(drop=True)

# Display the results
print(results_df)

# Save the results
results_df.to_csv("SVR_Hyperparameter_Results.csv", index=False)

print("Results saved successfully!")
```

```python
# Import the Random Forest Regression algorithm
from sklearn.ensemble import RandomForestRegressor

# Import ParameterGrid to generate all possible hyperparameter combinations
from sklearn.model_selection import ParameterGrid

# Import R2 Score to evaluate the model
from sklearn.metrics import r2_score

# Import Pandas to store the results
import pandas as pd

# --------------------------------------------------
# Define the hyperparameter combinations
# --------------------------------------------------

# param_grid = {

#     # Number of trees in the forest
#     "n_estimators": [50, 100, 200],

#     # Maximum depth of each tree
#     "max_depth": [None, 10, 20],

#     # Minimum samples required to split a node
#     "min_samples_split": [2, 5],

#     # Minimum samples required at a leaf node
#     "min_samples_leaf": [1, 2],

#     # Number of features considered for the best split
#     "max_features": ["sqrt", "log2"]

# }

param_grid = {
    "n_estimators": [200, 300],
    "criterion": ["squared_error", "friedman_mse"],
    "max_depth": [8, 10, 12],
    "min_samples_split": [2, 3],
    "min_samples_leaf": [2, 3],
    "max_features": ["sqrt"],
    "bootstrap": [True, False]
}

# Empty list to store the results
results = []

# --------------------------------------------------
# Hyperparameter Tuning
# --------------------------------------------------
for params in ParameterGrid(param_grid):

    # Create the Random Forest model
    model = RandomForestRegressor(
        random_state=42,
        **params
    )

    # Train the model
    model.fit(x_train, y_train)

    # Predict the test dataset
    y_pred = model.predict(x_test)

    # Calculate the R2 Score
    score = r2_score(y_test, y_pred)

    # Store the results
    results.append({
        "Algorithm": "Random Forest Regression",
        **params,
        "R2 Score": score
    })

# --------------------------------------------------
# Convert the results into a DataFrame
# --------------------------------------------------
results_df = pd.DataFrame(results)

# Sort by R2 Score
results_df = results_df.sort_values(
    by="R2 Score",
    ascending=False
).reset_index(drop=True)

# Display the results
print(results_df)

# Save the results to CSV
results_df.to_csv(
    "RandomForest_Hyperparameter_Results.csv",
    index=False
)

print("Results saved successfully!")
```
